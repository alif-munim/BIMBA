```python
from pathlib import Path
from sagemaker.pytorch import PyTorch
import sagemaker

role = sagemaker.get_execution_role()

SRC_DIR   = Path.cwd()                               # current dir = BIMBA
REQ_FILE  = SRC_DIR / "BIMBA-LLaVA-NeXT" / "requirements.txt"

estimator = PyTorch(
    entry_point="inference_script.py",               # file in SRC_DIR
    source_dir=str(SRC_DIR),                         # upload everything in BIMBA/
    dependencies=[str(REQ_FILE)],                    # extra pip requirements
    role=role,
    instance_type="ml.g5.12xlarge",
    instance_count=1,
    framework_version="2.0",
    py_version="py310",
    base_job_name="bimba-inference",
    environment={
        'VIDEO_S3': 's3://echodata25/test/1.2.276.0.7230010.3.1.2.1714512485.1.1703113078.10354201/1.2.276.0.7230010.3.1.3.1714512485.1.1703113078.10354202/mp4/1.2.276.0.7230010.3.1.4.859333938.1.1703114560.8025183.mp4'
    },
    disable_profiler=True,
)

estimator.fit(wait=True)
```
